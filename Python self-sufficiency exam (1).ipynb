1. Function to reverse a string:

```python
def reverse(word):
    new_word = str()
    for char in word:
        new_word = new_word + word[-1]
        word = word[0:-1]
    return new_word

reverse('forward')
```

```python
def bayes(a, b, b_on_a):
    post_a=(b_on_a*a)/(b)
    return post_a
```

3. Lahman database:

```python
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
from numpy import mean
from scipy import stats
from numpy.random import seed
sns.set()

teams=pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/lahman-teams.csv')
teams=teams.assign(SBperG = (teams.SB/teams.G))
teams_since_1995 = teams['yearID']>=1995
teams_we_want = teams.iloc[teams_since_1995.values]
ax=sns.distplot(teams_we_want['SBperG'])
```

```python
ax=sns.distplot(teams_we_want['W'])
```

```python
ax=sns.distplot(teams_we_want['ERA'])
```

The variables all seem to be fairly normal.

```python
sns.scatterplot(x='HR', y='W', data=teams_we_want)
teams_we_want['HR'].corr(teams_we_want['W'])
```

There is a significant correllation between how many home runs a team hits and how many games they win.

4. Samples of means of previous variables:

```python
teams_we_want['SBperG'].sample(n=100)

means = [mean(teams_we_want['SBperG'].sample(n=100)) for _ in range(1000)]


plt.hist(means)
plt.title('Stolen Bases per Game')
plt.show()
```

```python
teams_we_want['ERA'].sample(n=100)

means = [mean(teams_we_want['ERA'].sample(n=100)) for _ in range(1000)]


plt.hist(means)
plt.title('Earned Run Average')
plt.show()
```

The samples create a normal distribution.
